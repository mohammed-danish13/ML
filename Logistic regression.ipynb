```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['Department', 'salary']

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

encoded_features = encoder.fit_transform(df[categorical_cols])

encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_cols))

df_processed = pd.concat([df.drop(columns=categorical_cols), encoded_df], axis=1)

display(df_processed.head())
```

```python
X = df_processed.drop('left', axis=1)
y = df_processed['left']

print(f"Shape of features (X): {X.shape}")
print(f"Shape of target (y): {y.shape}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000, random_state=42, solver='liblinear')

model.fit(X_train, y_train)

print("Logistic Regression model trained successfully!")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
def predict_employee_turnover(model, encoder, employee_data):

    single_employee_df = pd.DataFrame([employee_data])

    categorical_sample = single_employee_df[categorical_cols]
    numerical_sample = single_employee_df.drop(columns=categorical_cols)

    encoded_features_sample = encoder.transform(categorical_sample)
    encoded_df_sample = pd.DataFrame(encoded_features_sample, columns=encoder.get_feature_names_out(categorical_cols))

    processed_sample = pd.concat([numerical_sample.reset_index(drop=True), encoded_df_sample], axis=1)

    missing_cols = set(X_train.columns) - set(processed_sample.columns)
    for c in missing_cols:
        processed_sample[c] = 0
    processed_sample = processed_sample[X_train.columns]

    prediction = model.predict(processed_sample)
    prediction_proba = model.predict_proba(processed_sample)[:, 1]

    return prediction[0], prediction_proba[0]

hypothetical_employee = {
    'satisfaction_level': 0.4,
    'last_evaluation': 0.5,
    'number_project': 2,
    'average_montly_hours': 160,
    'time_spend_company': 3,
    'Work_accident': 0,
    'promotion_last_5years': 0,
    'Department': 'sales',
    'salary': 'low'
}

predicted_class, predicted_probability = predict_employee_turnover(model, encoder, hypothetical_employee)

print(f"Hypothetical employee data: {hypothetical_employee}")
print(f"Predicted class (0=stays, 1=leaves): {predicted_class}")
print(f"Probability of leaving: {predicted_probability:.4f}")

if predicted_class == 1:
    print("This hypothetical employee is predicted to leave the company.")
else:
    print("This hypothetical employee is predicted to stay with the company.")
```
